```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('BDI_DATA_SET.csv')
```

```python
df['BDITotaScore'] = pd.to_numeric(df['BDITotaScore'], errors='coerce')
```

```python
conditions = [
    (df['BDITotaScore'] <= 13),
    (df['BDITotaScore'] >= 14) & (df['BDITotaScore'] <= 19),
    (df['BDITotaScore'] >= 20) & (df['BDITotaScore'] <= 28),
    (df['BDITotaScore'] >= 29)
]
choices = ['no depression', 'mild', 'moderate', 'high/severe']
df['depression_level'] = np.select(conditions, choices, default='no depression')
```

```python
X = df[['BDI1', 'BDI2', 'BDI3', 'BDI4', 'BDI5', 'BDI6', 'BDI7', 'BDI8', 'BDI9', 'BDI10',
        'BDI11', 'BDI12', 'BDI13', 'BDI14', 'BDI15', 'BDI16', 'BDI17', 'BDI18', 'BDI19', 'BDI20', 'BDI21']]
```

```python
df['has_depression'] = np.where(df['depression_level'] == 'no depression', 'no depression', 'has depression')
```

```python
y_stage1 = df['has_depression']
```

```python
X_train1, X_test1, y_train1, y_test1 = train_test_split(X, y_stage1, test_size=0.2, random_state=42)
```

```python
param_grid_stage1 = {
    'max_depth': [10, 20, 30, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 5],
    'criterion': ['gini', 'entropy'],
    'class_weight': ['balanced']
}

grid_search_stage1 = GridSearchCV(DecisionTreeClassifier(random_state=42), param_grid_stage1, cv=5, scoring='accuracy', n_jobs=-1)
grid_search_stage1.fit(X_train1, y_train1)
```

```python
dt_model_stage1 = grid_search_stage1.best_estimator_
y_pred_stage1 = dt_model_stage1.predict(X_test1)
```

```python
print("Stage 1 - No Depression vs. Has Depression")
print("Best Parameters:", grid_search_stage1.best_params_)
print("Accuracy:", accuracy_score(y_test1, y_pred_stage1))
print("\nClassification Report:\n", classification_report(y_test1, y_pred_stage1))
```

```python
cm_stage1 = confusion_matrix(y_test1, y_pred_stage1)
cm_normalized_stage1 = cm_stage1.astype('float') / cm_stage1.sum(axis=1)[:, np.newaxis]
plt.figure(figsize=(8, 6))
sns.heatmap(cm_normalized_stage1, annot=True, fmt=".2%", cmap="Blues", xticklabels=['no depression', 'has depression'], yticklabels=['no depression', 'has depression'])
plt.title("Normalized Confusion Matrix for Stage 1 - No Depression vs. Has Depression")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
X_stage2 = X[y_stage1 == 'has depression']
y_stage2 = df['depression_level'][y_stage1 == 'has depression']
```

```python
X_train2, X_test2, y_train2, y_test2 = train_test_split(X_stage2, y_stage2, test_size=0.2, random_state=42)
```

```python
param_grid_stage2 = {
    'max_depth': [10, 20, 30, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 5],
    'criterion': ['gini', 'entropy'],
    'class_weight': ['balanced', None],
    'max_features': [None, 'sqrt', 'log2']
}

grid_search_stage2 = GridSearchCV(DecisionTreeClassifier(random_state=42), param_grid_stage2, cv=5, scoring='accuracy', n_jobs=-1)
grid_search_stage2.fit(X_train2, y_train2)
```

```python
dt_model_stage2 = grid_search_stage2.best_estimator_
```

```python
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train2, y_train2)
```

```python
dt_model_stage2.fit(X_train_res, y_train_res)
y_pred_stage2_res = dt_model_stage2.predict(X_test2)
```

```python
print("Stage 2 - Severity Levels for Has Depression Group (with SMOTE)")
print("Best Parameters:", grid_search_stage2.best_params_)
print("Accuracy:", accuracy_score(y_test2, y_pred_stage2_res))
print("\nClassification Report:\n", classification_report(y_test2, y_pred_stage2_res))
```

```python
cm_stage2 = confusion_matrix(y_test2, y_pred_stage2_res, labels=y_stage2.unique())
cm_normalized_stage2 = cm_stage2.astype('float') / cm_stage2.sum(axis=1)[:, np.newaxis]
plt.figure(figsize=(8, 6))
sns.heatmap(cm_normalized_stage2, annot=True, fmt=".2%", cmap="Blues", xticklabels=y_stage2.unique(), yticklabels=y_stage2.unique())
plt.title("Normalized Confusion Matrix for Stage 2 - Depression Severity Prediction (SMOTE)")
plt.xlabel("Predicted Severity Level")
plt.ylabel("Actual Severity Level")
plt.show()
```

```python
cv_scores = cross_val_score(dt_model_stage2, X_stage2, y_stage2, cv=5, scoring='accuracy')
print(f'Cross-validation accuracy: {cv_scores.mean()}')
```

```python
cv_scores_res = cross_val_score(dt_model_stage2, X_train_res, y_train_res, cv=5, scoring='accuracy')
print(f'Cross-validation accuracy after resampling: {cv_scores_res.mean()}')
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

# Prepare features for association rule mining
X_bdi = df[['BDI1', 'BDI2', 'BDI3', 'BDI4', 'BDI5', 'BDI6', 'BDI7', 'BDI8', 'BDI9', 'BDI10',
            'BDI11', 'BDI12', 'BDI13', 'BDI14', 'BDI15', 'BDI16', 'BDI17', 'BDI18', 'BDI19', 'BDI20', 'BDI21']]

# Preprocess BDI responses into binary format (1 if score > 0, 0 otherwise)
binary_bdi = X_bdi.applymap(lambda x: 1 if x > 0 else 0)

# Add depression levels for association rule mining
binary_bdi['depression_level'] = df['depression_level']

# Filter the data to consider only rows with 'has depression'
binary_bdi_has_depression = binary_bdi[binary_bdi['depression_level'] != 'no depression']

# Apply Apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(binary_bdi_has_depression.drop('depression_level', axis=1),
                            min_support=0.2, use_colnames=True, max_len=3)

# Generate association rules from the frequent itemsets with a lift threshold of 0.5
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.5)

# Sort rules by lift to find the strongest rules
rules = rules.sort_values(by='lift', ascending=False)

# Display the top 10 rules based on lift
print("Top 10 Association Rules based on Lift:")
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(10))

# Plotting the top 10 association rules
plt.figure(figsize=(10, 6))
sns.barplot(x='lift', y='antecedents', data=rules.head(10), palette='viridis')
plt.title("Top 10 Association Rules Sorted by Lift")
plt.xlabel("Lift")
plt.ylabel("Antecedents")
plt.show()
```
